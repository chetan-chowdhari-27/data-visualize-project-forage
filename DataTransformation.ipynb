## **Data Loading**

```python
df_orders = spark.read.format("delta")\
            .option("inferSchema",True)\
            .option("header",True)\
            .load("abfss://5e53069f-335b-4b81-9294-0643a23d2e55@onelake.dfs.fabric.microsoft.com/a0643614-c093-4e00-9d36-582b677c372a/Tables/deltatable")
```

```python
display(df_orders)
```

## **Data Cleanup**

- ### Create a check that the quantity should not be below 1 unit - ### Create a check that the Unit price should not be below $0

```python
count_quanity_negative = df_orders[df_orders["Quantity"] < 0]
count_quanity_negative.count()
```

```python
count_unitprice_negative = df_orders[df_orders["UnitPrice"] < 0]
count_unitprice_negative.count()
```

```python
df_final = df_orders[(df_orders["Quantity"] < 0) & (df_orders["UnitPrice"] < 0)]
```

## **Checking If any Negative value**

```python
display(df_final[(df_final["Quantity"] < 0) & (df_final["UnitPrice"] < 0)])
```

### **Saving Data Enrich Table**

```python
df_final.write.format("delta")\
        .mode("append")\
        .saveAsTable("order_retail_final")
```
